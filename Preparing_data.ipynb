```python
import os
import xml.etree.ElementTree as ET
import random
from tqdm import tqdm
import shutil
```

```python
anotation_path = "/Users/thanhdat/Documents/Năm 3/Nhận dạng mẫu/Yolov8_Dog_Cat_Detection/archive 18.02.50/annotations"
label_pạth = "/Users/thanhdat/Documents/Năm 3/Nhận dạng mẫu/Yolov8_Dog_Cat_Detection/archive 18.02.50/labels"

classes = ["dog", "cat"]

for filename in os.listdir(anotation_path):
    tree = ET.parse(os.path.join(anotation_path, filename))
    root = tree.getroot()
    
    img_wid = int(root.find('size').find('width').text)
    img_hei = int(root.find('size').find('height').text)
    
    
    yolo_lines = []
    for obj in root.findall('object'):
        class_name = obj.find('name').text
        
        class_id = classes.index(class_name)
        bndbox = obj.find('bndbox') 
        xmin = int(bndbox.find('xmin').text)
        ymin = int(bndbox.find('ymin').text)
        xmax = int(bndbox.find('xmax').text)
        ymax = int(bndbox.find('ymax').text)    
        
        x_center = (xmin + xmax) / 2 / img_wid
        y_center = (ymin + ymax) / 2 / img_hei
        width = (xmax - xmin) / img_wid
        height = (ymax - ymin) / img_hei    
        
        yolo_line = f"{class_id} {x_center} {y_center} {width} {height}"
        yolo_lines.append(yolo_line)
        
    label_filename = os.path.splitext(filename)[0] + '.txt'
    with open(os.path.join(label_pạth, label_filename), 'w') as f:
        for line in yolo_lines:
            f.write(line + '\n')
```

```python
image_path = "/Users/thanhdat/Documents/Năm 3/Nhận dạng mẫu/Yolov8_Dog_Cat_Detection/archive 18.02.50/images"
output_path = "/Users/thanhdat/Documents/Năm 3/Nhận dạng mẫu/Yolov8_Dog_Cat_Detection/Data"
train_ratio = 0.8
val_ratio = 0.1
test_ratio = 0.1

for split in ["train", "val", "test"]:
    os.makedirs(os.path.join(output_path, "images", split), exist_ok=True)
    os.makedirs(os.path.join(output_path, "labels", split), exist_ok=True)
    
image_files = [f for f in os.listdir(image_path) if f.endswith(('.jpg', '.png'))]
image_files = [f for f in image_files if os.path.exists(os.path.join(label_pạth, os.path.splitext(f)[0] + ".txt"))]

random.shuffle(image_files)

n_total = len(image_files)
n_train = int(train_ratio * n_total)
n_val = int(val_ratio * n_total)
n_test = n_total - n_train - n_val


train_files = image_files[:n_train]
val_files = image_files[n_train:n_train + n_val]
test_files = image_files[n_train + n_val:]



        
def copy_files(files, split):
    for file in tqdm(files, desc=f"Copying {split}"):
        name = os.path.splitext(file)[0]
        img_src = os.path.join(image_path, file)
        label_src = os.path.join(label_pạth, name + ".txt")

        img_dst = os.path.join(output_path, "images", split, file)
        label_dst = os.path.join(output_path, "labels", split, name + ".txt")

        shutil.copy(img_src, img_dst)
        shutil.copy(label_src, label_dst)
        
        
copy_files(train_files, "train")
copy_files(val_files, "val")
copy_files(test_files, "test")

```
